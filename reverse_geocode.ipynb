```python
import reverse_geocoder as rg
```

```python
import csv
import pandas as pd

d = pd.read_csv('./data/wildfires_10k_sample.csv')
```

```python
coords = zip(d['LATITUDE'], d['LONGITUDE'])
coords
```

```python
places = rg.search(coords)
```

```python
places[0]
```

```python
len(d['OBJECT_ID'])
```

```python
d_new = []

for object_id, place in zip(d['OBJECT_ID'], places):
    d_new.append([str(object_id), place['name'], place['admin2'], place['admin1'], place['cc']])
```

```python
cols = ['OBJECT_ID', 'CITY', 'COUNTY', 'STATE', 'COUNTRY']
d_new_df = pd.DataFrame(d_new, columns = cols)
```

```python
d_new_df.to_csv('./data/wildfire_places.csv', columns=cols)
```
